```python
import pandas as pd
import numpy as np
import os
import rasterio
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from tqdm import tqdm

# === CONFIG ===
LABEL_FILE = "labeled.csv"
IMAGE_DIR = "labeled_samples"
CLASSES = ["LIVE", "DEAD", "BARE"]
REPEATS = 5
K_RANGE = range(10, 26)
HOLDOUT_YEAR = 2022  # For single-year testing

# === LOAD FULL LABELED DATA ===
df = pd.read_csv(LABEL_FILE)
df = df[df["label"].isin(CLASSES)]

# === HELPER: Load pixel and compute NDVI ===
def extract_pixel_bands_with_ndvi(tif_path):
    with rasterio.open(tif_path) as src:
        data = src.read()
        red, green, blue, nir = data[0, 0, 0], data[1, 0, 0], data[2, 0, 0], data[3, 0, 0]
        ndvi = (nir - red) / (nir + red + 1e-6)
        return [red, green, blue, nir, ndvi]

# === PART A: Combined-Year Training + 80/20 Random Split
results_random = []

print("\n📊 Running 80/20 random split across all years...\n")

for k in tqdm(K_RANGE):
    accs = []

    for rep in range(REPEATS):
        X, y = [], []

        dfs = []
        for cls in CLASSES:
            cls_df = df[df["label"] == cls].sample(n=k, random_state=rep)
            dfs.append(cls_df)
        subset = pd.concat(dfs).sample(frac=1, random_state=rep).reset_index(drop=True)

        for _, row in subset.iterrows():
            tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
            if os.path.exists(tif_path):
                X.append(extract_pixel_bands_with_ndvi(tif_path))
                y.append(row["label"])

        X = np.array(X)
        y = np.array(y)

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, random_state=rep, stratify=y
        )

        clf = RandomForestClassifier(n_estimators=100, random_state=rep)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accs.append(accuracy_score(y_test, y_pred))

    results_random.append({
        "k": k,
        "mean_acc": np.mean(accs),
        "std_acc": np.std(accs)
    })

# === PART B: Combined-Year Training + Single-Year Test
results_holdout = []

print(f"\n📊 Running generalization test: Train on all years, test on {HOLDOUT_YEAR}...\n")

df_holdout = df[df["year"] == HOLDOUT_YEAR]
df_rest = df[df["year"] != HOLDOUT_YEAR]

for k in tqdm(K_RANGE):
    accs = []

    for rep in range(REPEATS):
        X_train, y_train = [], []

        dfs = []
        for cls in CLASSES:
            cls_df = df_rest[df_rest["label"] == cls].sample(n=k, random_state=rep)
            dfs.append(cls_df)
        train_subset = pd.concat(dfs).sample(frac=1, random_state=rep).reset_index(drop=True)

        for _, row in train_subset.iterrows():
            tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
            if os.path.exists(tif_path):
                X_train.append(extract_pixel_bands_with_ndvi(tif_path))
                y_train.append(row["label"])

        # Prepare test set
        X_test, y_test = [], []
        for _, row in df_holdout.iterrows():
            tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
            if os.path.exists(tif_path):
                X_test.append(extract_pixel_bands_with_ndvi(tif_path))
                y_test.append(row["label"])

        if len(X_test) == 0:
            print(f"⚠️ No test data found for {HOLDOUT_YEAR}. Skipping...")
            continue

        X_train = np.array(X_train)
        X_test = np.array(X_test)
        y_train = np.array(y_train)
        y_test = np.array(y_test)

        clf = RandomForestClassifier(n_estimators=100, random_state=rep)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accs.append(accuracy_score(y_test, y_pred))

    results_holdout.append({
        "k": k,
        "mean_acc": np.mean(accs),
        "std_acc": np.std(accs)
    })

# === PLOT COMPARISON ===
k_vals = [r["k"] for r in results_random]
random_means = [r["mean_acc"] for r in results_random]
random_stds = [r["std_acc"] for r in results_random]

holdout_means = [r["mean_acc"] for r in results_holdout]
holdout_stds = [r["std_acc"] for r in results_holdout]

plt.figure(figsize=(10, 6))
plt.errorbar(k_vals, random_means, yerr=random_stds, fmt='-o', label='80/20 Random Split', capsize=4)
plt.errorbar(k_vals, holdout_means, yerr=holdout_stds, fmt='-s', label=f'Test on {HOLDOUT_YEAR}', capsize=4)
plt.title("Accuracy vs Labeled Samples per Class (NDVI Included)")
plt.xlabel("Labeled Samples per Class (k)")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import os
import rasterio
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from tqdm import tqdm

# === CONFIG ===
LABEL_FILE = "labeled.csv"
IMAGE_DIR = "labeled_samples"
CLASSES = ["LIVE", "DEAD", "BARE"]
YEARS = [2014, 2016, 2018, 2020, 2022]
K_RANGE = range(10, 26)
REPEATS = 5

# === LOAD DATA ===
df = pd.read_csv(LABEL_FILE)
df = df[df["label"].isin(CLASSES)]

# === NDVI FEATURE FUNCTION ===
def extract_pixel_bands_with_ndvi(tif_path):
    with rasterio.open(tif_path) as src:
        data = src.read()
        red, green, blue, nir = data[0, 0, 0], data[1, 0, 0], data[2, 0, 0], data[3, 0, 0]
        ndvi = (nir - red) / (nir + red + 1e-6)
        return [red, green, blue, nir, ndvi]

# === GRAPH 1: Accuracy vs k for each test year (trained on full data) ===
per_year_results = {}

print("\n📊 Running fixed full-dataset training + per-year testing...\n")

for test_year in YEARS:
    df_test = df[df["year"] == test_year]

    per_k_accs = []

    for k in tqdm(K_RANGE, desc=f"Testing on {test_year}"):
        accs = []

        for rep in range(REPEATS):
            # Sample k per class from full dataset (all years)
            X_train, y_train = [], []

            for cls in CLASSES:
                cls_df = df[df["label"] == cls].sample(n=k, random_state=rep)
                for _, row in cls_df.iterrows():
                    tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
                    if os.path.exists(tif_path):
                        X_train.append(extract_pixel_bands_with_ndvi(tif_path))
                        y_train.append(row["label"])

            # Prepare test set from the target year
            X_test, y_test = [], []
            for _, row in df_test.iterrows():
                tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
                if os.path.exists(tif_path):
                    X_test.append(extract_pixel_bands_with_ndvi(tif_path))
                    y_test.append(row["label"])

            if len(X_train) == 0 or len(X_test) == 0:
                continue

            clf = RandomForestClassifier(n_estimators=100, random_state=rep)
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            accs.append(accuracy_score(y_test, y_pred))

        per_k_accs.append((k, np.mean(accs)))

    per_year_results[test_year] = per_k_accs

# === PLOT GRAPH 1 ===
plt.figure(figsize=(10, 6))
for year, accs in per_year_results.items():
    k_vals, year_accs = zip(*accs)
    plt.plot(k_vals, year_accs, marker='o', label=f"Test on {year}")

plt.title("Accuracy per Year (Training on Full Dataset)")
plt.xlabel("Labeled Samples per Class (k)")
plt.ylabel("Accuracy")
plt.ylim(0.4, 1.0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === GRAPH 2: Accuracy vs k on 80/20 Split ===
try:
    k_vals = [r["k"] for r in results_random]
    random_means = [r["mean_acc"] for r in results_random]
    random_stds = [r["std_acc"] for r in results_random]

    plt.figure(figsize=(8, 5))
    plt.errorbar(k_vals, random_means, yerr=random_stds, fmt='-o', label='80/20 Random Split', capsize=4)
    plt.title("Accuracy vs Labeled Pixels per Class (NDVI Included, 80/20 Split)")
    plt.xlabel("Labeled Samples per Class (k)")
    plt.ylabel("Accuracy")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

except NameError:
    print("⚠️ 'results_random' not found. Please run the 80/20 experiment first.")
```

```python
import pandas as pd
import numpy as np
import os
import rasterio
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

# === CONFIG ===
LABEL_FILE = "labeled.csv"
IMAGE_DIR = "labeled_samples"
CLASSES = ["LIVE", "DEAD", "BARE"]
YEARS = [2014, 2016, 2018, 2020, 2022]
K = 25  # number of samples per class used for training

# === LOAD LABEL DATA ===
df = pd.read_csv(LABEL_FILE)
df = df[df["label"].isin(CLASSES)]

# === NDVI Feature Extraction ===
def extract_pixel_bands_with_ndvi(tif_path):
    with rasterio.open(tif_path) as src:
        data = src.read()
        red, green, blue, nir = data[0, 0, 0], data[1, 0, 0], data[2, 0, 0], data[3, 0, 0]
        ndvi = (nir - red) / (nir + red + 1e-6)
        return [red, green, blue, nir, ndvi]

# === Confusion Matrices ===
fig, axs = plt.subplots(1, len(YEARS), figsize=(20, 4))
all_reports = {}

for i, year in enumerate(YEARS):
    df_test = df[df["year"] == year]
    df_train = df

    # Sample K per class from entire dataset for training
    X_train, y_train = [], []
    for cls in CLASSES:
        cls_df = df_train[df_train["label"] == cls].sample(n=K, random_state=42)
        for _, row in cls_df.iterrows():
            tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
            if os.path.exists(tif_path):
                X_train.append(extract_pixel_bands_with_ndvi(tif_path))
                y_train.append(row["label"])

    # Prepare test set for this year
    X_test, y_test = [], []
    for _, row in df_test.iterrows():
        tif_path = os.path.join(IMAGE_DIR, row["filename"].replace(".png", ".tif"))
        if os.path.exists(tif_path):
            X_test.append(extract_pixel_bands_with_ndvi(tif_path))
            y_test.append(row["label"])

    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    cm = confusion_matrix(y_test, y_pred, labels=CLASSES)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CLASSES)
    disp.plot(ax=axs[i], colorbar=False)
    axs[i].set_title(f"Test on {year}")

    # Save classification report for summary table
    report = classification_report(y_test, y_pred, labels=CLASSES, output_dict=True)
    all_reports[year] = report

plt.suptitle("Confusion Matrices (k=25 per class, Trained on Full Dataset)", fontsize=14)
plt.tight_layout()
plt.show()

# === Summary Table (Accuracy + F1s) ===
summary_stats = {
    "Year": [],
    "Accuracy": [],
    "F1_LIVE": [],
    "F1_DEAD": [],
    "F1_BARE": []
}

for year in YEARS:
    report = all_reports[year]
    summary_stats["Year"].append(year)
    summary_stats["Accuracy"].append(report["accuracy"])
    summary_stats["F1_LIVE"].append(report["LIVE"]["f1-score"])
    summary_stats["F1_DEAD"].append(report["DEAD"]["f1-score"])
    summary_stats["F1_BARE"].append(report["BARE"]["f1-score"])

df_summary = pd.DataFrame(summary_stats)
print(df_summary.round(3))
```
